```python
iris = sns.load_dataset('iris')
```

```python
correlation = iris.corr()
```

```python
correlation
```

```python
sns.heatmap(iris.corr() , cmap='coolwarm' , annot=True )
```

```python
sns.set( style = 'whitegrid' ,color_codes=True)
```

```python
sns.lmplot(x = 'sepal_length' , y= 'petal_length' ,hue='species' ,
          data=iris,palette="Set1", markers=['o' ,'s' , 'D'])
```

```python
sns.lmplot( x='sepal_length', y='petal_length' , hue='species' ,
          data=iris , fit_reg=False, palette='Set1' ,markers=['o' , 's' , 'D'])
```

```python


sns.pairplot(iris, height=5, vars=["sepal_length", "petal_length"], 
             
             markers=["o", "s", "D"], hue="species")

```

```python
sns.pairplot(iris, markers= ['o' , 's' , 'D']  ,hue='species')
```

```python
sns.pairplot(iris , kind='reg' , hue='species', markers=['o', 's' , 'D'], palette='Set1')
```

```python
sns.boxplot(data=iris , orient='h')
```

```python
iris['ID'] = iris.index
iris['ratio'] = iris['sepal_length']/iris['sepal_width']
sns.lmplot(x = 'ID' , y='ratio' ,
          data=iris , hue='species' ,markers=['o','s','D'])
```

```python
sns.stripplot( x= 'species' , y= 'sepal_length' , data= iris)
```

```python
sns.stripplot(x = 'species' , y='petal_length' , data=iris)
```

```python
sns.swarmplot(x = 'species' , y= 'petal_length' , data=iris)
```
